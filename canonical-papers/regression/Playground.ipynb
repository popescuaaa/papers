### Simple and multivariate regression type

```python
from lr import SimpleLinearRegression
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

# Configure fig size
plt.style.use('ggplot')
```

```python
dataset = load_diabetes()
```

```python
X = dataset.data[:, 0] # Take only the first parameter from that dataset
y = dataset.target
print(f"X: {X.shape}, y: {y.shape}")
```

```python
plt.scatter(X, y)
plt.figure(figsize=(8, 8))
plt.show()
```

```python
lr = SimpleLinearRegression(x = X, y = y)
lr.normalize()
lr.fit()


print(f"Params before training: w: {lr.weight}, b: {lr.bias}")
loss_history = lr.train(iterations=100, learning_rate=3.1)
print(f"Params after training: w: {lr.weight}, b: {lr.bias}")

# Plot the loss history over the number of iterations
plt.plot(loss_history)
```

```python
x_test = lr.x_test
y_test = lr.y_test
predictions = lr.predict(x_test)

# Plot predictions over real values
plt.scatter(x_test, y_test)
plt.plot(x_test, predictions)
plt.show()
```
